###  Time Series and Sentiment Analysis in Facebook Groups

Time series analysis is useful for analyzing how frequently a topic  is being discussed - and whether discussion on a topic is increasing or decreasing over a period of time. Sentiment analysis is useful for analyzing whether discussion related to a given topic is positive or negative (or neutral).* When combined, time series and sentiment analysis provide insight into whether discussion is increasing on a given topic during a fixed period -- and whether that discussion is trending positively or negatively . 

**There are many different sentiment analysis machine learning models -- all of which are trained in different ways on different datasets. All sentiment analysis modles have inherent limitations and weaknesses, e.g. difficulty detecting sarcasm, evaluating complex sentences, identifying targets of speech, etc. Consequently, sentiment analysis outputs should be treated with some skepticism. However, at a minimum, sentiment analysis is useful for identifying overall trends in sentiment. Moreover, they are also very useful as a mechanism to filter content that may be problematic, i.e. speech classified as 'negative' may potential hateful speech or dangerous speech that requires action. Where possible, you should manually review the outputs before publishing your findings!*

In this example, we're going to perform time series and sentiment analysis on messages from posts in the community of pro-Georgia Republican Facebook Groups we collected in *3. Analyzing Media Sources and Articles in Facebook Groups* and cleaned in *4. Extracting Narrative Topics in Facebook Groups*. This example will enable us to demonstrate what is possible as well as the limitations of sentiment analysis. 

First, we import packages needed for script and add a database connection to collect posts from our community of Facebook groups.

```python
import pandas as pd
import numpy as np
import nltk
import nltk.corpus
from datetime import datetime
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import pyodbc 
from sqlalchemy import create_engine
import re
import networkx as nx
import matplotlib.pyplot as plt
import string
from textblob import TextBlob
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

connection_string = "your_connection_string"
engine = create_engine(connection_string, encoding='utf8')
db = engine.connect()
```

From our database, we'll grab the data from pro-Georgia Republican Facebook groups that we cleaned and stored in our previous notebook *4. Extracting Narrative Topics in Facebook Groups*

```python
ga_rep_group_posts = pd.read_sql_table("ga_republican_groups_posts_clean", con=db)
```

We'll pass the message text that we cleaned in the previous notebook through a sentiment analysis algorithm that classifies sentiment on a probability scale from -1 (most negative) to 1 (most positive). We'll create a new column in our dataset called 'probability' in which to hold the probability score.     

```python
def sentimentAnalysis(text):
    return TextBlob(text).polarity  

ga_rep_group_posts['probability'] = ga_rep_group_posts['text'].apply(sentimentAnalysis)
```

Now that we have our scores, we'll create a new column called 'sentiment' that assign a label to the text according to its probability score:

* If the probability score is above 0, label sentiment as 'positive'
* If the probability score is below 0, label sentiment as 'negative'
* If the probability score is exactly 0, label sentiment as 'neutral'

**Alternatively, we could use a more nuanced approach to labelling that includes 'neutral' sentiment if the score is between -0.3 and +0.3. For the purposes of this example, we'll use a simple and straightfoward approach (however code for a nuanced labelling approach is included below and commented out).*

```python
def sentimentLabel(row):
    if row['probability'] < 0: #for nuanced labelling, change to < -0.3
        val = 'negative'
    elif row['probability'] > 0: #for nuanced labelling, change to > 0.3
        val = 'positive'
    else:
        val = 'neutral'
    return val

ga_rep_group_posts['sentiment'] = ga_rep_group_posts.apply(sentimentLabel, axis=1)
```

We'll take a quick peek at our data to see how some of the text has been scored and labeled:

```python
ga_rep_group_posts[['sentiment', 'probability', 'message']].sort_values(by='probability', ascending=False)
```

At cursory glance, it looks like our sentiment analyzer did a decent job of characterizing positive and negative messages (at least highly positive and highly negative)

Let's plot the amount of positive and negative messages by date. 

First we'll extract the date (year-month-day) from the column in our dataset that contains both date and timestamp of the post.

```python
ga_rep_group_posts['date'] = pd.to_datetime(ga_rep_group_posts['date'])
ga_rep_group_posts['ymd'] = pd.to_datetime(ga_rep_group_posts['date']).dt.to_period('D') #extract just y/m/d
```

Then we'll write a function called 'dateSentiment' to group and sum the posts by date and sentiment.

```python
def dateSentiment(data):
    temp = pd.DataFrame(data[['sentiment', 'ymd']])
    temp = pd.DataFrame(temp.groupby(['ymd','sentiment']).size())
    return temp.pivot_table(index ='ymd', columns='sentiment', values=0, fill_value=0, aggfunc='sum').unstack()

ga_rep_groups_sentiment = dateSentiment(ga_rep_group_posts)
```

Now we'll create a function called 'plotSentiment' that we can use to display how much content is positive or negative. 

```python
def plotSentiment(data):
    date = data.index.get_level_values('ymd').unique().to_timestamp()
    line_pos = data.loc[(data.index.get_level_values('sentiment') == "positive")].values
    line_neg = data.loc[(data.index.get_level_values('sentiment') == "negative")].values
    plt.plot_date(date, line_pos, 'g')
    plt.plot_date(date, line_neg, 'r')
    plt.xlabel('Date')
    plt.ylabel("Post Count")
    plt.xticks(date.get_level_values('ymd').unique(), rotation=90)
    plt.show()

plotSentiment(ga_rep_groups_sentiment)
```

Overall, it looks like there are more positive messages than negative messages being posted. However, it would be more interesting to know what is the content of positive and negative messages. 

We'll create a wordcloud to see if we can drill down into the topics of positive and negative messages.

First we'll look at which words are most prominent in positive messages (first we'll remove some words we know are unlikely to be informative):

```python
stop_words.update(['Georgia', 'State', 'County', 'Time', 'PG', 'GA', 'live', 'people'])

np.random.seed(2020)
positive_messages = ga_rep_group_posts[ga_rep_group_posts['sentiment'] == 'positive']
positive_text = " ".join(word for word in positive_messages['text'])
wordcloud = WordCloud(stopwords=stop_words, background_color="white", max_words=40).generate(positive_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Looks like there's a lot of positive messages related to the election, voting, and Trump (and mentions of special session, vote red, and patriot).

Now let's look at which words are most prominent in negative messages.

```python
np.random.seed(2020)
negative_messages = ga_rep_group_posts[ga_rep_group_posts['sentiment'] == 'negative']
negative_text = " ".join(word for word in negative_messages['text'])
wordcloud = WordCloud(stopwords=stop_words, background_color="white", max_words=40).generate(negative_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Here we find negative messages related the election, voting, and Trump. We also see democrat mentioned prominently (and mentions of biden, fraud, and warnock -- the only Georgia Senate candidate whose name appears in either wordcloud!) 

We can drill down how positively or negatively specific topics that are being discussed by extracting just the messages in which keywords of interest appear. 

For example, we can extract mentions of 'warnock' in our text and plot a graph that shows how positive or negative he is being discussed.  

```python
warnock_posts = ga_rep_group_posts[ga_rep_group_posts['text'].str.contains('warnock')]
warnock_posts_sentiment = dateSentiment(warnock_posts)
plotSentiment(warnock_posts_sentiment)
```

Although it's a small sample size, it looks like the posts mentioning Warnock are quite mixed in terms of sentiment, although the sentiment is more negative on average than in all posts. 

We'll take a peek at the individual posts to see the actual text of both positive and negative messages mentioning 'warnock', starting with the negative posts:  

```python
all_warnock_posts = warnock_posts[['sentiment', 'probability', 'message', 'text']].sort_values(by='probability', ascending=True)
all_warnock_posts[all_warnock_posts['sentiment'] == 'negative'].sort_values(by='probability', ascending=True).head(15)
```

It looks like these messages are relatively negative. Clearly the sentiment analyzer focused on words such as 'fake', 'trouble' and 'loathe' to facilitate classification as 'negative'. 

Now let's take a look at the positive posts.

```python
all_warnock_posts[all_warnock_posts['sentiment'] == 'positive'].sort_values(by='probability', ascending=False).head(15)
```

These posts are not particularly positive...it is likely the sentiment analyzer picked up words such as 'perfect', 'beautiful', and 'true' as signals of positivity. The message 'Yep, he's putting on an act...' is clearly not positive and should have been classified as negative!

This illustrates the need for manual review of sentiment analysis outputs. On the upside, the sentiment analyzer noted that mentions of Warnock broadly trended more negatively than posts overall. However, the lack of precision of individual findings is problematic!

Finally, we'll create two wordclouds to highlight what words are most prominent in posts that mention 'warnock' labelled as 'negative' or 'positive'. We'll start with 'negative' posts:

```python
stop_words.update(['warnock'])
np.random.seed(2020)
warnock_negative_messages = all_warnock_posts[all_warnock_posts['sentiment'] == 'negative']
warnock_negative_text = " ".join(word for word in warnock_negative_messages['text'])
wordcloud = WordCloud(stopwords=stop_words, background_color="white", max_words=40).generate(warnock_negative_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

The wordcloud of 'negative' posts is more insightful and highlights key lines of attack used against Warnock, i.e. that he is anti-Israel, his pro-choice stance undermines his Christian faith, etc.

Let's look at words in posts labelled as 'positive':

```python
np.random.seed(2020)
warnock_positive_messages = all_warnock_posts[all_warnock_posts['sentiment'] == 'positive']
warnock_positive_text = " ".join(word for word in warnock_positive_messages['text'])
wordcloud = WordCloud(stopwords=stop_words, background_color="white", max_words=40).generate(warnock_positive_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

This is not terribly insightful, but certainly is a contrast to the 'negative' posts wordcloud. Mentions of 'Perdue', 'Loeffler' and 'vote' may indicate that messages attacking Warnock also contained positive messaging aimed at getting out the vote for Republican candidates.
